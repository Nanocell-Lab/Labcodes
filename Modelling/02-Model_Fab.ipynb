```python
def cys_distances(file_pdb,chain):
    """Si las distancia es superior a 5 A no estn enlazadas"""
    import prody
    import itertools
    
    prot1 = prody.parsePDB(file_pdb,chain = chain)
    
    cysteines = [] 
    for res in prot1.iterResidues():
        if res.getResname() == "CYS":
            cysteines.append(res)
    print(cysteines)
    for cys1, cys2 in itertools.combinations(cysteines, 2):
        print(cys1,cys2)
        atom1 = cys1.getAtom("SG")
        print(atom1)
        atom2 = cys2.getAtom("SG")
        print(atom2)
        atom3 = cys1.getAtom("CA")
        atom4 = cys2.getAtom("CA")
        if prody.calcDistance(atom3, atom4) < 5:
            print (cys1, cys2, "%4.2f"%prody.calcDistance(atom1, atom2), "%4.2f"%prody.calcDistance(atom3, atom4))
        else:
            pass
            print("Not cysteines")
    return
```

## Make model with multiple templates (require malign) 

```python
# Comparative modeling with multiple templates
from modeller import *              # Load standard Modeller classes
from modeller.automodel import *    # Load the automodel class



# directories for input atom files
#env.io.atom_files_directory = ['.', '../atom_files']

# Redefine the special_patches routine to include the additional disulfides
# (this routine is empty by default):
class MyModel(automodel):
    def special_patches(self, aln):
        # A disulfide between residues 8 and 45:
        self.patch(residue_type='DISU', residues=(self.residues['22'],
                                                  self.residues['45']))
log.verbose()    # request verbose output
env = environ()  # create a new MODELLER environment to build this model in

a = MyModel(env,
              alnfile  = 'toalign_FabC2.fas.mafft', # alignment filename
              knowns   = ('1ngzB_FabC1_oriented','1yntB_1_14_53','3hotA_DC1_oriented'),     # codes of the templates
              sequence = 'Fab_C2_dimer')               # code of the target
a.starting_model= 1                 # index of the first model
a.ending_model  = 3
#a.md_level = refine.very_slow # index of the last model
                                    # (determines how many models to calculate)
a.make()                            # do the actual comparative modeling
```
